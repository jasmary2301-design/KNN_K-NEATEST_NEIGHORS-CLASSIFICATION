**KNN -K NEAREST NEIGHBORS**

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

**DATA SET LOADING**

```python
Dataset = load_iris()
Dataset
```

```python
X = Dataset.data
y = Dataset.target
```

```python
import pandas as pd
df = pd.DataFrame(X,columns=Dataset.feature_names)
```

```python
df
```

**TARGET**

```python
y
```

```python
print(set(y))
Dataset.target_names
```

```python
df["target"]=y
df
```

**TEST TRAIN SPLIT**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X_train
```

**SCALER**

```python
std_scaler = StandardScaler()
mm_scaler = MinMaxScaler()
X_train_std = std_scaler.fit_transform(X_train)
X_train_mm = mm_scaler.fit_transform(X_train)
```

```python
X_train_std
```

```python
X_train_mm
```

```python
X_test = mm_scaler.transform(X_test)
```

```python
X_test
```

```python
Flowers = KNeighborsClassifier(n_neighbors= 5)
Flowers.fit(X_train_mm,y_train)
```

```python
y_pred = Flowers.predict(X_test)
```

```python
print("ACCURACY = ",accuracy_score(y_test,y_pred))
print("\n\nCLASSIFICATION REPORT :\n\n",classification_report(y_test,y_pred))
print("\n\nCONFUSION MATRIX :\n\n",confusion_matrix(y_test,y_pred))
```

```python
print("ACCURACY = ",accuracy_score(y_test,y_pred))
print("\n\nCLASSIFICATION REPORT :\n\n",classification_report(y_test,y_pred))
print("\n\nCONFUSION MATRIX :\n\n",confusion_matrix(y_test,y_pred))
```

**VISUALIZATION**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
CFM= confusion_matrix(y_test,y_pred)
labels=Dataset.target_names


plt.figure(figsize=(8,4))

sns.heatmap(CFM,annot=True,fmt="d",xticklabels=labels,yticklabels=labels,cmap="Blues")

plt.title("CONFUSION MATRIX - HEATMAP")
plt.xlabel("PREDICTED CONDITION")
plt.ylabel("ACTUAL CONDITION")
plt.show()
```
